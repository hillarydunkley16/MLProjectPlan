```python

features_drop = ['fnlwgt', 'education', 'relationship', 'capital-gain', 'capital-loss']
df =df.drop(features_drop, axis = 1)
df.columns
```

```python
print(df.columns)
corr_matrix = round(df.corr(),5)
corr_matrix
```

```python

nan_count = np.sum(df.isnull(), axis = 0)
nan_detected = nan_count != 0 
#print(nan_detected) 
is_int_or_float = (df.dtypes == 'int64')| (df.dtypes == 'float64')
to_impute = nan_detected & is_int_or_float
print(to_impute) 
cols_to_impute = list(df.columns[to_impute])
print(cols_to_impute)
```

```python
for x in cols_to_impute: 
    name = x+"_na"
    df[name] = df[x].isnull()
print(df.columns)
```

```python
for x in cols_to_impute: 
    df[x].fillna(value = df[x].mean(), inplace = True)
print(df.head(10))
```

```python
df.rename(columns = {'income_binary' : 'income_above50k?'}, inplace = True)
df.columns
```

```python
df.replace({'income_above50k?': {'>50K': True, '<=50K': False}}, inplace = True)
df.head(10)
```

```python
feature_list = [x for x in df if x != 'income_above50k?']
#print(feature_list)
df_onehot = [x for x in feature_list if df[x].dtype == 'object']
print(df_onehot)
```

```python
df['workclass'].fillna(value = "workclass_unavailable", inplace = True)
#df['workclass'].unique()
df["workclass"] = df["workclass"].replace(["Without-pay", "Never-worked"], "No-income")
#df['workclass'].head()
df['workclass'].unique()
```

```python
df_workclass = pd.get_dummies(data = df['workclass'])
df_workclass.index = df.index 
```

```python
df = pd.concat([df, df_workclass], axis = 1)
df = df.drop('workclass', axis = 1)
```
